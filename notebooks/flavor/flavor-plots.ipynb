```python
import json
import pickle
import os

import matplotlib
matplotlib.rcParams.update({'font.size': 18})

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

from coffea import hist
```

```python
year = '2018'
if not os.path.isdir(year):
    os.mkdir(year)
    
pickle_path = '../../output/pickle/vhbb_v5/'+year+'/ParticleNet_msd.pkl'
```

```python
h = pickle.load(open(pickle_path,'rb')).integrate('region','signal').integrate('systematic','nominal')
h.identifiers('process')
```

```python
vh = h.integrate('process',['WH','ZH']).sum('bb1')
```

```python
ggf = h.integrate('process','ggF').sum('bb1')
```

```python
vbf = h.integrate('process','VBFDipoleRecoilOn').sum('bb1')
```

```python
tth = h.integrate('process','ttH').sum('bb1')
```

```python
vv = h.integrate('process','VV').sum('bb1')
```

```python
zjets = h.integrate('process','Zjets').sum('bb1')
```

```python
wjets = h.integrate('process','Wjets').sum('bb1')
```

```python
ttbar = h.integrate('process','ttbarBoosted').sum('bb1')
```

```python
singlet = h.integrate('process','singlet').sum('bb1')
```

```python
labels = ['unmatched','light','charm','bottom']
```

```python
plt.imshow(ggf.sum('genflavor1','genflavor2').values()[()].transpose(),extent=[40,201,40,201],origin='lower')
plt.suptitle('ggF')
plt.xlabel(r'Jet 1 $m_{sd}$')
plt.ylabel(r'Jet 2 $m_{sd}$')
plt.savefig(year+'/ggf_msd1_msd2.pdf',bbox_inches='tight')
```

```python
hist.plot1d(ggf.sum('msd2','genflavor2'),overlay='genflavor1',stack=True)
plt.legend(labels=labels)
plt.suptitle('ggF')
plt.savefig(year+'/ggf_msd1.pdf',bbox_inches='tight')
```

```python
hist.plot1d(ggf.sum('msd1').sum('genflavor1'),overlay='genflavor2',stack=True)
plt.suptitle('ggF')
plt.savefig(year+'/ggf_msd2.pdf',bbox_inches='tight')
```
